4.) Matrix Multiplication

Rule:

1.) The inner dimensions must  match:

```(3,2) @ (3,2)``` won't work

```(2,2) @ (2,3)``` will work

2.) The resulting matrix has the shape of the outer dimension

```(2,3) @ (3,2) ``` -> ```(2,2)```



```Important point to remember ```

Last two dimensions are always treated as matrices.

All dimensions before the last two are batch dimensions.

1D Ã— 1D â†’ dot product â†’ scalar output.

2D Ã— 2D â†’ normal matrix multiplication.

2D Ã— 1D â†’ matrixâ€“vector multiplication.

1D Ã— 2D â†’ vectorâ€“matrix multiplication.

If any tensor has â‰¥3 dimensions, batched matrix multiplication is used.

Matrix multiplication requires inner dimensions to match.

Batch dimensions must be broadcastable.

Batch dimensions are compared from right to left.

Two batch dimensions are compatible if they are equal or one is 1.

Broadcasting reuses data; it does not copy it.

Output batch shape is the broadcasted batch shape.

Output matrix shape is (n Ã— p).

1D tensors are temporarily expanded and then squeezed back.

Dot-product (1D Ã— 1D) does not support out=.

matmul automatically chooses the correct multiplication type.

````python
# vector x vector
tensor1 = torch.randn(3)
tensor2 = torch.randn(3)
print(torch.matmul(tensor1, tensor2).size())
# matrix x vector
tensor1 = torch.randn(3, 4)
tensor2 = torch.randn(4)
print(torch.matmul(tensor1, tensor2).size())
# batched matrix x broadcasted vector
tensor1 = torch.randn(10, 3, 4)
tensor2 = torch.randn(4)
print(torch.matmul(tensor1, tensor2).size())
# batched matrix x batched matrix
tensor1 = torch.randn(10, 3, 4)
tensor2 = torch.randn(10, 4, 5)
print(torch.matmul(tensor1, tensor2).size())
# batched matrix x broadcasted matrix
tensor1 = torch.randn(10, 3, 4)
tensor2 = torch.randn(4, 5)
print(torch.matmul(tensor1, tensor2).size())
````

````python
# Shapes need to be in the right way
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]], dtype=torch.float32)

tensor_B = torch.tensor([[7, 10],
                         [8, 11],
                         [9, 12]], dtype=torch.float32)

torch.matmul(tensor_A, tensor_B) # (this will error)
````

````python
# View tensor_A and tensor_B
print(tensor_A)
print(tensor_B)
````

````python
print(tensor_A)
print(tensor_B.T)
````

````python
# The operation works when tensor_B is transposed
print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}\n")
print(f"New shapes: tensor_A = {tensor_A.shape} (same as above), tensor_B.T = {tensor_B.T.shape}\n")
print(f"Multiplying: {tensor_A.shape} * {tensor_B.T.shape} <- inner dimensions match\n")
print("Output:\n")
output = torch.matmul(tensor_A, tensor_B.T)
print(output)
print(f"\nOutput shape: {output.shape}")
````

````python
torch.mm(tensor_A,tensor_B.T)
````

#### Reshaping, stacking, squeezing and unsqueezing

````python
x = torch.arange(1.,9.)
x , x.shape , x.ndim, x
````

If we use reshape and manipulate the new tensor and previous tensor , both will remain independent (creates deep copy of the original data variable)

````python
x_reshaped = x.reshape(1,2,4)
x_reshaped,x_reshaped.shape,x_reshaped.ndim,x_reshaped
````

But in  the view it creates the new view of the same tensor(creates shallow copy of original data variable)

````python
z = x.view(1,8)
z, z.shape,z.ndim
````

````python
z[:,0] = 5
z,x
````

````python
x_stacked = torch.stack([x,x,x,x],dim=0)
x_stacked
````

````python
x_stacked = torch.stack([x,x,x,x],dim=1)
x_stacked
````

````python
x_squeezed = x_reshaped.squeeze()
x_squeezed,x_squeezed.shape
````

````python
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
x_unsqueezed,x_unsqueezed.shape
````

````python
x_unsqueezed = x_squeezed.unsqueeze(dim=2)
x_unsqueezed,x_unsqueezed.shape
````

ðŸ”¥ Fun Face :

Because permuting returns a view (shares the same data as the original), the values in the permuted tensor will be the same as the original tensor and if you change the values in the view, it will change the values of the original

````python
x_original = torch.rand(size=(224,224,3))
print(x_original,x_original.shape)
x_permuted = x_original.permute(2,0,1)
print("_______-----")
print(x_permuted.shape,x_permuted)
````
