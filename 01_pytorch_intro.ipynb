# Creating tensors

#### 1.)Scaler
A scalar is a single number and in tensor it is a zero dimension tensor

````python
scalar = torch.tensor(9)
scalar
````

````python
scalar.ndim,scalar.dtype
````

````python
scalar.item(),scalar.item
````

#### 1.)Vector
A vector is a single dimension tensor but can contain many numbers where the scalar can have one number.

````python
vector = torch.tensor([4,7])
vector
````

````python
vector.ndim
````

##### ðŸ”¥Fun fact :
To tell the number of dimensions a tesnor in pytorch have count the number of square brackets on the outside (```[```) but only need to count one side.

````python
vector.shape
````

#### 3.)Matrix


````python
MATRIX = torch.tensor(
    [
      [4,7],
      [9,2]
    ]
    )
MATRIX
````

````python
MATRIX.ndim,MATRIX.dtype
````

#### Creating the random tensors

````python
random_tensor = torch.rand(size=(3,4))
````

````python
random_tensor
````

````python
random_image_size_tensor = torch.rand(size=(224,224,3))
random_image_size_tensor.shape,random_image_size_tensor.ndim,random_image_size_tensor.dtype
````

````python
random_image_size_tensor
````

````python
# creating a tensor of all zeros
zeros = torch.zeros(size=(3,4))
zeros,zeros.dtype
````

````python
# creating a tesnor of all ones
ones = torch.ones(size=(3,4))
ones,ones.dtype,ones.dim(),ones.shape
````

````python
zero_to_ten = torch.arange(0,10,1)
````

````python
zero_to_ten
````

````python
# copying the shape of the a tensor
ten_zeros =  torch.zeros_like(input=zero_to_ten)
ten_zeros
````

````python
ten_ones = torch.ones_like(input=zero_to_ten)
ten_ones
````

````python
float_32_tensor = torch.tensor([3.0,6.0,9.0],
                               dtype=None,
                               device='cpu',
                               requires_grad=False)
````

````python
float_32_tensor.shape,float_32_tensor.dtype,float_32_tensor.device
````
